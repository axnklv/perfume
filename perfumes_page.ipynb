```python
import pandas as pd
import requests
from lxml import html
```

```python
url = 'https://www.fragrantica.ru/designers-1/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) \
            Chrome/39.0.2171.95 Safari/537.36'}
```

```python
page = requests.get(url, headers=headers)
data = html.fromstring(page.content)
```

```python
page.status_code
```

```python
aromas_list = data.xpath('//div[@class="nduList"]')
```

```python
for item in aromas_list[0].text_content().split('\n'):
    if item != '':
        print(item.strip())
```

```python
for catalog in data.xpath('//div[@class="nduList"]')[0:]:
    print(catalog.text_content().replace('\n',''))
```

```python
def getDatafromPage(data):
    voc = {}
    for count,catalog in enumerate(data.xpath('//div[@class="nduList"]')[0:]):
        for p in catalog.xpath('p'):
            A = {}
            for a in p.xpath('a'):
                A['url'] = a.attrib.get('href')
                A['name'] = a.text_content().replace('\n',' ')
            voc[count] = A
    return pd.DataFrame.from_dict(voc, orient ='index')
```

```python
page = getDatafromPage(data)
```

```python
df = page.copy()
df.head()
```

```python
df['letter'] = df['name'].str.split(' ').str[1]
```
